# Introduction of HEAP

Heap is a data structure which we will implement using array in CSE220 course.

We know of two catagories of heap, namely max heap nad min heap. characteristics of these:

1. Max heap have largest value on top root
2.min heap have smallest value on top root.



In order to make a heap the array has to form a complete binanry tree. Complete binary tree is a tree structure where leaf nodes have no child. If a node has only one child then the child is left child only.

# functions to access parent, left child and right child for index idx :

left child's index is 2*parents index

```python
def left(idx):
    return 2*idx
```

right childs index is 2*parents index +1

```python
def right(idx):
    return 2*idx+1
```

to get the parent index of a child floor divide the childs index by 2. however is the childs index is 1, i.e the top root of our tree, we shouldn't get his parent.

```python
def parent(idx):
    if idx ==1:
        return None
    else:
        return idx//2
```

to know the size of heap use this. size of a heap is the number of valid entities in the array. a valid entity is any thing but None.

the parameter `(heap)` here is the adress of the array. This array has indexing staring from 0. We will ignore 0 index in this code.

```python
def heap_size(heap):
    size =0
    for entity in range(1 , len (heap)):
        if heap[entity] != None:
            size +=1
    return size
```

Example of heap_size iterative function working

```python
heap = [None, 10, 15, 20, None, 25]
print(heap_size(heap))
```

```python
def heap_size(heap, index=1):
    if index >= len(heap) or heap[index] is None:
        return 0

    left_size = heap_size(heap, 2 * index)
    right_size = heap_size(heap, 2 * index + 1)

    return 1 + left_size + right_size
```

Example of heap_size recursive function working

```python
heap = [None, 10, 15, 20, None, 25]
print(heap_size(heap))
```
